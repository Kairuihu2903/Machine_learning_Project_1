# Purge Dict of numbers and symbols

```python
import re
regex = re.compile('.$')



# the function below will search for all . or $ in a given string and append that to dot_or_doll list
def run(string):
    dot_or_dol = []
    # Make own character set and pass
    # this as argument in compile method
    regex = re.compile('[.$]')
     
    # Pass the string in search
    # method of regex object.   
    if(regex.search(string) == None):
      pass
        #print("String is accepted", j)
         
    else:
      return(string)
# the function below will search if string has numbers or special characters in a given string and append that to dot_or_doll list
def run2(string):
    hasnum = []
    # Make own character set and pass
    # this as argument in compile method
    regex = re.compile('[1234567890!@#$%^&*<>?:;`~+/*-=]')
     
    # Pass the string in search
    # method of regex object.   
    if(regex.search(string) == None):
      pass
        #print("String is accepted", j)
         
    else:
      return(string)

#function below will return cleane(no . or $) dict
def getcleanishare(etf_name):
  keystochange = []
  changedkey = []
  saved_dict = getiShareHoldings(etf_name)
  nonumdic = saved_dict.copy()
 

  # below will search though all keys in getiShareHolding and then seperate the ones with . or $

  for j in saved_dict.keys():
    if run2(str(j)) != None:
      nonumdic.pop(run2(str(j)))
  notlongdict = nonumdic.copy()
  for j in nonumdic.keys():
    if len(j) > 4:
      notlongdict.pop(j)


  for j in notlongdict.keys():
    if run(str(j))!= None:
      keystochange.append(run(str(j)))


  # bellow will replace . or $ with dot_or_dol
  for i in keystochange:
    my_string = i
    my_new_stringifdot = re.sub('[.$]', 'dot_or_dol', my_string)
    changedkey.append(my_new_stringifdot)

  count = 0
# bellow will replace the changed key with the new key
  for i in changedkey:
    notlongdict[str(i)] = notlongdict.pop(str(keystochange[count]))
    count += 1

  notlongdict_2 = notlongdict.copy()
# below will stringify every key in the dict then replace the old key
  for i in notlongdict:
    notlongdict_2[str(i)] = notlongdict_2.pop(i)

  return(notlongdict_2)


getcleanishare("EEM")
```

# Jaccard stuff

```python
# See https://en.wikipedia.org/wiki/Jaccard_index
def jaccard(set1,set2):
  if len(set1.union(set2)) != 0:
    return len(set1.intersection(set2))/len(set1.union(set2))



jaccard(set(getlistincol("IVV")), set(getlistincol("SPY")))
#jaccard( set(classDb['XLK'].dropna().index),set(classDb['QQQ'].dropna().index) )
```

```python
#program will take a ETF name and if valid then it will iterate thought the list of etfs in the database
# the the program will run jaccard and compare the user selected ETF with every other ETF in the list
# and place the ETF name and the jaccard result into a dict.
# we then iterate though that dict to get the ETF most similar to the the selected ETF

def get_recommend():
  ETF = input("what ETF would you like to purchase: ")
  count = 0
  ETFlen = len(inDb())
  ETFexist = 0 # 0 == false 1 == true
  jdict = {}
  largest = 0
  name = []
  for i in inDb():
    count += 1
    if ETF == i and ETFexist == 0:
      print("your ETF is", ETF)
      ETFexist = 1
    elif count == ETFlen and ETFexist == 0:
      return("the ETF", ETF, "does not exist in the database")  
  if ETFexist == 1:
    for i in inDb():
      jdict[i] = {"similar" : jaccard(set(getlistincol(str(ETF))), set(getlistincol(str(i))))}

  for i in jdict:
    if jdict[i]['similar'] > largest and i != ETF:
      largest = jdict[i]['similar']
      name = i
  return("we would recommend purchasing ",name," because it has a ", largest*100, "% similarity to ", ETF)







print(get_recommend())
#print(inDb())
```
